```python
#import the libraries
import pandas as pd
import scipy
import numpy as np
from scipy import stats
```

# **LABTAT**

```python
df = pd.read_csv("LabTAT.csv")
```

```python
df
```

```python
# Perform one-way ANOVA test
from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(df['Laboratory 1'], df['Laboratory 2'], df['Laboratory 3'], df['Laboratory 4'])
```

```python
# Output results
print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

```python
df1
```

```python
unitA=pd.Series(df1.iloc[:,0])
unitA
```

```python
unitB=pd.Series(df1.iloc[:,1])
unitB
```

**Cutlets**

```python
df2 = pd.read_csv("Costomer+OrderForm.csv")
```

```python
df2
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Performing chi-square test for independence
chi2_stat, p_val, dof, expected = chi2_contingency(pd.crosstab(df2['Phillippines'], [df2['Indonesia'], df2['Malta'], df2['India']]))
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
```
